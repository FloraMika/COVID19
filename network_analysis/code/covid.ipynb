**Summary**  
The objective was to achieve a systematic and comprehensive characterization of the cell response to virus. For this, we performed a weighted co-expression network analysis on both the transcriptomic and proteomic dataset. This allowed us to respectively identify a set of 5 and 4 communities of strongly interconnected genes and proteins, and further characterize them funtionally. Among all communities we identified several pathways of interest including HIF-1, mTOR, and TNF signaling. Further, we ranked all communities based on their centrality to identify those that tend to display a higher number of central genes, i.e. those potentially influencing network behavior more strongly. These communities are associated with SLC2A1, AKT1, SEC13, RAF1, and CASP8. Further, they tend to be enriched in mTOR and MAPK signaling, Lysosomal and Proteasome-related processes, as well as cell cycle control. Finally, we aimed to identify the most similar communities at the proteo-transcriptomic level and found statistically significant functional association between 3 transcriptomic and 3 proteomic communities. Interestingly, among the associated communities, genes and proteins tend to be associated with mTOR signaling, among other processes.

***Preprocessing***  
RNAseq filtered by removing genes with median RPKM < 1. Total: 5594 genes expressed and considered for further analysis. Proteomic removed non-infected sample 3 since it was clustering at 48hours.

***Network analysis***  
Association analyses performed by computing Spearman rank correlations and filtered for FDR < 0.01. At this significance level, all correlations show abs(R) > 0.71. A weighted network analysis was performed on the set of positive correlations identified from RNAseq (5532 genes, ~764,000 edges, R>0.83, FDR <0.01) and proteomic (7517 proteins, ~1.3m edges, R>0.85, FDR < 0.01). Similar analyses were performed on negative networks (where Spearman rho was multiplied by -1) but are not further explored. All networks were compared against randomly generated networks  (Erdos Renyi model) with same size and edge number and degree distributions were compared, among other metrics including density, average path length, and diameter. 
Overall, all networks display markedly distinct features than those displayed by the random networks. Among the top 5% most central (i.e. connected, and thus potentially important) genes and proteins were found several membrane adaptor proteins involved in intracellular membrane fusion (AP1B1, AP2A1, AP2S1, AP3D1), calpains (CAPN1), G-associated proteins (GNB2), protein-translocation and signaling (SSR4m TRMT1, WDR5), morphogenesis (TMEM11), and metabolism (G6PD). Among these, G6PD was found at the top 1% most central genes and proteins.

***Community analysis***  
A community analysis was then performed on each of the networks through the Leiden algorithm by considering modularity optimization and disregarding communities with fewer than 30 elements. Respectively 5 and 4 communities were identified in the RNAseq (728 - 1355 nodes/community) and proteomic network (1361 - 2519 nodes/community). The RNAseq community shows 2 substantially more central communities (mean normalized degree ~0.09) compared to other communities in the same network (mean normalized degree <0.036). In turn, the proteomic network displays 3 communities with approximate centralities (~normalized degree ~0.061 - 0.07) and one that tends to be less connected (mean normalized degree <0.003).

***Functional analysis***  
GSEA was then applied on each network for KEGG pathways 2019 through Enricher, where background was considered as the full set of quantified genes and proteins, respectively. 
Among all statistically significant processes (FDR < 0.05), we observe several associated with complement and coagulation cascades, HIF-1, mTOR, TNF and ErbB signaling. Other 30 and 33 Gene Ontology biological processes and KEGG terms associated with viral infection or response to infection were respectively identified in the RNAseq and proteomic networks. Several other processes are found in both networks including

- *Amino acid metabolism:* Amino sugar and nucleotide sugar metabolism; Arg and Pro metabolism; Glycine, serine and threonine metabolism; Valine, leucine and isoleucine degradation  
- *Sugar Metabolism:* Glycolysis / Gluconeogenesis; Glyoxylate and dicarboxylate metabolism; Citrate cycle (TCA cycle); Oxidative phosphorylation; Propanoate metabolism; Pyruvate metabolism; N-Glycan biosynthesis; One carbon pool by folate
- *Lipid metabolism:* Fatty acid degradation; Fatty acid elongation; Steroid biosynthesis; Synthesis and degradation of ketone bodies; Terpenoid backbone biosynthesis
- *Signaling:* FoxO signaling pathway; Longevity regulating pathway; Parathyroid hormone synthesis, secretion and action; PPAR signaling pathway; Thermogenesis; Complement and coagulation cascades


***Network equivalence***  
To identify the most similar communities at the proteo-transcriptomic level we computed hypergeometric tests for the overlap between functional terms (FDR <0.05) between both communities. This was done by considering enriched pathways, and not by overlap of genes/proteins, since the coverage of the two datasets (RNAseq and proteomic) is very different. While analyses were done considering Gene Ontology BPs, KEGG terms and Human Gene Atlas terms, the following results pertain to the analysis of community similarity.  
Overall, a direct equivalence between communities in both networks is not easily identified, and in some cases there are communities that are exclusive from one network but show no statistically significant similarity with any community in the opposing network.  
Communities 1 and 5 in the RNAseq network, and community 3 in the proteomic network, show no statistically significant (FDR > 0.05) association with any of the opposing networks. In turn, communities 2 - 4 in the RNAseq network show statistically significant (FDR < 0.008) associations with communities 1, 2 and 4 of the proteomic network. This suggests functional similarity shown by some of the communities in the two networks. Such an overlap is substantially increased if one considers Gene Ontology biological processes where no communities become *"isolated"* (FDR < 0.05, results not shown).   
Interestingly, oxidative phosphorylation, thermogenesis, mTOR signaling, and folate metabolism are shared terms among statistically associated communities at the proteo-transcriptomic level.  
No genes of interest, associated with KEGG terms of interest, are identified in both networks. While this seems like a negative result it is not surprising for the following reasons. First, the two omics contain distinct coverage, and a different number of samples (11 and 12) were used to build the two networks, thus leading to distinct networks. Second,  successive cleaning of the data was performed in order to formulate the networks. Third, extremely conservative filtering criteria were employed by finding statistically significant associations between communities, followed by a selection of a subset genes (of interest) that are associated with a subset of pathways (of interest) that pass multiple statistical tests.

```python
def build_process_network(name, pos_neg, bkgd):
    my_ass=dict_aa[name]
    
    if(pos_neg=='pos'):
        my_ass=my_ass.loc[my_ass.R>0]
    elif(pos_neg=='neg'):
        # negative weights are converted to positive so that weighted networks can be used
        my_ass=my_ass.loc[my_ass.R<0]
        my_ass.R=my_ass.R*-1
    else:
        print('Problem')
    
    
    mynet=na.network_analysis(my_ass, main_dir=results+name+'/'+pos_neg+'/', output_name=name+'_network')
    ns=mynet.build_network(weighted=True)
    nc=mynet.centralities_compute().head()
    cf,cfs=mynet.community_detection(weighted=True)
    en=mynet.enrichment(organism='human', background=bkgd)
    return(ns,nc,cf,cfs,en)
```

```python
# from statsmodels.stats.multitest import multipletests

def get_functional_overlap(net_obj1,net_obj2,gene_set=None,FDRthr=0.05, FDR_method='fdr_bh',verbose=False):
    """
    Compares communities of two networks to find the most similar communities through hypergeometric testing.
    
    Total GSEA terms used as input from those passing significance in the 2 networks.
    
    """
    if(gene_set!=None):
        er1=net_obj1.enrichment_output.loc[net_obj1.enrichment_output.Gene_set==gene_set].copy()
        er2=net_obj2.enrichment_output.loc[net_obj2.enrichment_output.Gene_set==gene_set].copy()
    else:
        er1=net_obj1.enrichment_output.copy()
        er2=net_obj2.enrichment_output.copy()
        
    er1=er1.loc[er1['Adjusted P-value']<FDRthr]
    er2=er2.loc[er2['Adjusted P-value']<FDRthr]
    
    #total terms in the entire network that pass significance
    total_terms=pd.unique(np.append(er1['Term'].values,er2['Term'].values))
    
    #comms
    cc1_list=er1['comm'].unique()
    cc2_list=er2['comm'].unique()
    
    if(verbose):
        print('FDR chosen: ',str(FDRthr))
        print('# communities in network 1: ',str(len(cc1_list)))
        print('# communities in network 2: ',str(len(cc2_list)))
        print('# universe biological processes: ',str(len(total_terms)))
    
    overlap_terms=pd.DataFrame()

    for cc1 in cc1_list:
        for cc2 in cc2_list:
            df1=er1.loc[er1.comm==cc1,]
            df2=er2.loc[er2.comm==cc2,]
            
            #Gene Set intersection
            t1=df1.loc[:,'Term'].values
            t2=df2.loc[:,'Term'].values
            inter=np.intersect1d(t1,t2)
            hypergeom_c1c2=aa.hypergeom_list(total_terms, t1, t2)
            
            intersecting_genes=[]
            for iterm in inter:
                g1=df1.loc[df1.Term==iterm,'Genes'].values[0].split(';')
                g2=df2.loc[df2.Term==iterm,'Genes'].values[0].split(';')
                intersecting_genes=np.append(intersecting_genes,
                    ';'.join(np.intersect1d(g1,g2))
                                            )
                
            
            if(verbose):
                print(f'Comm1: {cc1} (n: {len(t1)}) | Comm2: {cc2} (n: {len(t2)}) | Intersection: {len(inter)} | P: {hypergeom_c1c2}')
            
            out=pd.Series(
                [cc1,cc2,hypergeom_c1c2,
                 len(inter),inter,t1,len(t1),t2,len(t2),intersecting_genes,len(intersecting_genes)],
                index=['Net1_comm','Net2_comm','P',
                       'n_inter','inter','Comm1_vals','n_Comm1_vals','Comm2_vals','n_Comm2_vals','intgenes_for_interterms','n_intgenes']
            )
            overlap_terms=pd.concat([overlap_terms, out],1)
    
    overlap_terms=overlap_terms.T.reset_index(drop=True)
    overlap_terms['FDR']=multipletests(overlap_terms['P'], method=FDR_method)[1]
    overlap_terms=overlap_terms.loc[overlap_terms.FDR<FDRthr]
    return(overlap_terms)

get_functional_overlap(rna_pos,prot_pos,gene_set='KEGG_2019_Human')
```

```python
#computes mean community centrality

def community_centrality(network_in, centrality_metric='degree'):
    all_metrics=[ 'degree', 'eccentricity', 'betweeness', 'closeness', 'deg_w',
       'btw_w', 'cls_w', 'degree_rank', 'betweeness_rank', 'closeness_rank',
       'eccentricity_rank']
    communities=network_in.community_feats.copy()
    icentrality=network_in.centralities_network
    
    my_comm_centralities=pd.DataFrame()
    for cc in communities.CC_comm_id:
        temp=communities.loc[communities.CC_comm_id==cc,'val'].values[0]
        
        metric_cen=[]
        for cen_i in all_metrics:
            
            temp_cen=icentrality.loc[temp,cen_i]
            node_centrality=np.mean(temp_cen)
            metric_cen=np.append(metric_cen,node_centrality)
            
        out=pd.Series(np.append([cc],metric_cen), index=np.append(['CC_comm_id'],['mean_'+x for x in all_metrics]))
        my_comm_centralities=pd.concat([my_comm_centralities, out],1)
        
    my_comm_centralities=my_comm_centralities.T
    my_comm_centralities.reset_index(inplace=True,drop=True)
    
    out=pd.merge(communities, my_comm_centralities, on='CC_comm_id')
    
    full_net_size=out['size'].sum()
    out['degree_norm']=out.apply(lambda x: np.float(x['mean_degree'])/full_net_size,1)
    return(out)

# community_centrality(rna_pos)
```
